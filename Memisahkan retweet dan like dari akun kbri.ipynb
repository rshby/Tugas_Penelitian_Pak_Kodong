```python
stock_file = sorted(glob("Gabungan/data_*.csv"))
stock_file
```

```python
df = pd.concat((pd.read_csv(file, sep=",", encoding='cp1252').assign(filename = file) for file in stock_file), ignore_index=True)
df.head()
```

```python
# melihat kolom
df.columns
```

```python
filt = ((df["lang"]=="in") | (df["lang"]=="en")) & (df["retweet_count"]>=1)
df = df.loc[filt][["user_id","screen_name","created_at","source", "location","text","is_retweet", "is_quote","favorite_count","retweet_count", "quote_count","lang"]].reset_index()
df.drop(columns="index", inplace=True)
```
