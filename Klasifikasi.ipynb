```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Datasets

```python
df = pd.read_csv("gabungan_file.csv")
df.drop(columns="Unnamed: 0", inplace=True)
```

### Fungsi Klasifikasi by text

```python
def Klasifikasi(mode):
    if mode == "A":
        filt = (df["text"].str.contains("WNI")) | (df["text"].str.contains("Dokumen")) | (df["text"].str.contains("#NegaraMelindungi")) | (df["text"].str.contains("#Covid19"))
    elif mode == "B":
        filt = (df["text"].str.contains("Hubungan Diplomatik")) | (df["text"].str.contains("Hubungan Bilateral")) | (df["text"].str.contains("Kerja sama")) | (df["text"].str.contains("#IniDiplomasi")) | (df["text"].str.contains("#SahabatKemlu")) | (df["text"].str.contains("#MenluRetno")) | (df["text"].str.contains("#FMMarsudi")) | (df["text"].str.contains("#RintisKemajuan")) | (df["text"].str.contains("#BatikDiplomacy")) | (df["text"].str.contains("#IndonesianWay")) | (df["text"].str.contains("#DigitalDiplomacy")) | (df["text"].str.contains("#DiplomasiKuliner"))
    elif mode == "C":
        filt = (df["text"].str.contains("UN / United Nation")) | (df["text"].str.contains("ASEAN")) | (df["text"].str.contains("Kerjasama Multilateral")) | (df["text"].str.contains("Rohingya")) | (df["text"].str.contains("Palestina")) | (df["text"].str.contains("Suriah")) | (df["text"].str.contains("Irak")) | (df["text"].str.contains("ISIS")) | (df["text"].str.contains("#IndonesiaUntukDunia")) | (df["text"].str.contains("#CounterTerrorism")) | (df["text"].str.contains("#DiplomacyforPeaceandProsperity"))
    elif mode == "D":
        filt = (df["text"].str.contains("Papua")) | (df["text"].str.contains("Sawit")) | (df["text"].str.contains("Perbatasan")) | (df["text"].str.contains("Demokrasi Indonesia")) | (df["text"].str.contains("#BaliDemocracyForum"))
        
    return df.loc[filt]
```

# Hasil Klasifikasi
### Klasifikasi: A

```python
print("Retweet:", Klasifikasi("A")["retweet_count"].sum(),", Like:", Klasifikasi("A")["favorite_count"].sum())
Klasifikasi("A").sort_values(by="retweet_count", ascending=False)
```

### Klasifikasi: B

```python
print("Retweet:", Klasifikasi("B")["retweet_count"].sum(),", Like:", Klasifikasi("B")["favorite_count"].sum())
Klasifikasi("B").sort_values(by="retweet_count", ascending=False)
```

### Klasifikasi: C

```python
print("Retweet:", Klasifikasi("C")["retweet_count"].sum(),", Like:", Klasifikasi("C")["favorite_count"].sum())
Klasifikasi("C").sort_values(by="retweet_count", ascending=False)
```

### Klasifikasi: D

```python
print("Retweet:", Klasifikasi("D")["retweet_count"].sum(),", Like:", Klasifikasi("D")["favorite_count"].sum())
Klasifikasi("D").sort_values(by="retweet_count", ascending=False)
```
